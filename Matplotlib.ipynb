# Matplotlib
* It is the most popular plotting library for python.
* It gives you control over every aspect of a figure.
* It was designed to have a similar feel to MatLab's  graphical plotting.

## Importing

```python
import matplotlib.pyplot as plt
%matplotlib inline # Used only in jupyter-notebook, so that we don't have to type 'plt.show()' again n again..
import numpy as np
import pandas as pd
```

## There are two ways of plotting through matplotlib:
1. Functional Method
2. Object Oriented Method

### 1. Functional Method

```python
# Creating data to plot
x = np.linspace(0,5,11)
y = x**2
```

```python
# plotting
plt.plot(x,y) # it's that easy
plt.xlabel('X Label') # labels x-axis
plt.ylabel('Y Label') # lables y-axis
plt.title('Title') # Gives a title
```

### subplots 

```python
# creating a subplot 
plt.subplot(1,2,1) # (row,column,the place you are selecting to plot)
plt.xlabel('X Label')
plt.ylabel('Y Label') 
plt.title('Title')
plt.plot(x,y,'r') # 'r' refers to color
plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.xlabel('X Label')
plt.ylabel('Y Label') 
plt.title('Title')
```

### 2. Object Oriented Method

```python
fig = plt.figure() # Think of it as a empty canvas
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # [left,bottom,width,height]
axes.plot(x,y,color='y')
axes.set_xlabel('X-Label')
axes.set_ylabel('Y-Label')
axes.set_title('Title');
```

### putting 2 plots on the same canvas

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y,color='r')
axes1.set_title('Larger Plot')
axes2.plot(y,x,color='b')
axes2.set_title('Smaller Plot'); # ';' hides the info above the plot
```

### subplots

```python
# Creating a subplot
fig,axes = plt.subplots(nrows=1,ncols=2) # (no. of rows, no. of cols)
```

```python
# another example...
fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout() # fixing the overlapping
```

```python
# finally...
fig,axes =  plt.subplots(nrows=1,ncols =2)
axes[0].plot(x,y,color='red')
axes[1].plot(y,x,color='orange')
```

## Figure Size and DPI

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,5),dpi=90) # dpi : dots per inch (default value is 100)
axes[0].plot(x,y,color='r')
axes[1].plot(y,x,color='y')
```

### saving a fig

```python
fig.savefig('figure.png',dpi=100) # saves the figure into png format (can also be converted into diff formats)
```

### legend

```python
fig = plt.figure(figsize=(6,4))
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,x**2,label='Squared')
axes.plot(x,x**3,label='Cubed')
fig.legend(loc=10) # Legend is added (loc: position of the legend)
```

## Appearance

```python
# color
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y,color='#202020') # color hex code given..
```

```python
# line-width,markers,alpha..
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y,color='#202020',linewidth=3,marker='o',markersize=10,markeredgecolor='black',markerfacecolor='red',alpha=0.5); # alpha means transperancy
```

```python
# x-limit,y-limit
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y,color='#202020',linewidth=3,marker='o',markersize=10,markeredgewidth=3,markeredgecolor='black',markerfacecolor='red',alpha=0.9)
ax.set_xlim([0,5]) # setting x-axis
ax.set_ylim([0,30]); # setting y-axis
```

```python
# Including everything we learnt till now and trying an example, let's see.....
fig = plt.figure(figsize=(6,4),dpi=100)
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,x**2,color='#202020',linewidth=3,marker='o',linestyle='dotted',markersize=10,markeredgewidth=3,markeredgecolor='black',markerfacecolor='red',label='X-Squared',alpha=0.8)
ax.plot(x,x**3,color='#202020',linewidth=3,marker='o',linestyle='dashed',markersize=10,markeredgewidth=3,markeredgecolor='black',markerfacecolor='yellow',label='X-Cubed',alpha=0.8)
ax.set_xlim([0,5]) # setting x-axis
ax.set_ylim([0,30])  # setting y-axis
ax.legend(loc='best'); # adding a legend
```
